Reading our images and converting to grayscale

```python
path = 'C:\Danyal\Work\FAST\Semester 5\Digital Image Processing\i192014_DanyalFaheem_19_A3\`data\img2.png'
img = cv2.imread(path)
```

```python
#Reading our image
img = cv2.imread(path)
#Thresholding our image to get a binary image
gray =  cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
thresh, out = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
#Finding the external contours in the image
contours, hierarchy = cv2.findContours(out,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
#Finding the first bounding Rectangle of the contours to get aspect ratio
x,y,w,h = cv2.boundingRect(contours[0])
aspectRatio = w/h
x1 = y1 = w1 = h1 = 0
#Looping through contours to find the smallest aspect ratio bounding rectange
for i in range(len(contours)):
    x,y,w,h = cv2.boundingRect(contours[i])
    #Finding the object with the smallest aspect ratio, this would inherently mean the highest frequency as waves would be shorter
    if aspectRatio < w/h:
        aspectRatio = w/h
        x1,y1,w1,h1 = x,y,w,h
#Drawing a rectangle on the highest frequency curve
cv2.rectangle(img,(x1,y1),(x1+w1,y1+h1),(0,255,0),2)

#Displaying our image
plt.imshow(img)
plt.axis('off')
```
