```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Reading our image

```python
path = 'C:\Danyal\Work\FAST\Semester 5\Digital Image Processing\i192014_DanyalFaheem_19_A3\`data\img3.png'
img = cv2.imread(path);
```

```python
#Reading our image
img = cv2.imread(path)
#Thresholding our image to get a binary image
gray =  cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
thresh, out = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
#Finding the external contours in the image
contours, hierarchy = cv2.findContours(out,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
#Finding the bounding rectangles of the first rectangle and the last rectangle
x, y, w, h = cv2.boundingRect(contours[0])
x1, y1, w1, h1 = cv2.boundingRect(contours[7])
#Using the starting position of the first bounding rectangle and the ending position of the last rectangle, drawing rectangle over it
cv2.rectangle(img, (x, y), (x1+w1, y1+h1), (0, 255, 0), 2)
#Printing the amount of contours - 1 to amount for the entire image as a contour too
cv2.putText(img, str(len(contours) - 1) , (100,100), 3, 3, 255)

#Displaying our image
fig, axs = plt.subplots(1, 1, figsize=(8, 8))

axs.imshow(img, cmap='gray')
axs.axis('off')


```
